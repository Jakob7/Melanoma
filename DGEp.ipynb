```python
rna_samples = pd.read_csv('data/gdc_sample_sheet.2018-08-09.tsv', sep = '\t')
rna_samples
```

```python
names = list(rna_samples['File Name'])
samples = list(rna_samples['Case ID'])
i = 0
for ever in rna_samples['File ID']:
	df = pd.read_csv('data//rawRNAdata//'+ever+"//"+names[i], sep = '\t', names = ['gene', 'count'])
	df.to_csv('data//rawRNAdata//'+samples[i]+ ".txt", sep = '\t', index = False)
	i += 1
```

```python
e = 0
DATA = []

for every in samples:
	df = pd.read_csv('data\\rawRNAdata\\'+every + '.txt', sep = "\t")
	if e == 0:
		genes = list(df['gene'])
		e +=1
	a = []
	counts = list(df['count'])
	a.append(every)
	a += counts
	DATA.append(a)
	del a

BIG_DATA = {"gene": genes}
DataFrame = pd.DataFrame(BIG_DATA, columns = ["gene"])
u = 0
for ever in DATA:
	DataFrame[ever[0]] = ever[1:]
	u+=1
	print(u)
DataFrame.to_csv("data//rawRNA.txt", sep = "\t", index = False)
```

```python
RAWrna = pd.read_csv('data/rawRNA.txt', sep = '\t')
maf = pd.read_csv('data/TCGA.maf', sep = '\t')
ZNF99names = list(maf[(maf['Hugo_Symbol'] == "ZNF99")&(maf['HGVSp'] == 'p.Ser3Leu')]['Tumor_Sample_Barcode'])
MAGEC1names = list(maf[(maf['Hugo_Symbol'] == "MAGEC1")&(maf['HGVSp'] == 'p.Glu620Lys')]['Tumor_Sample_Barcode'])
meta = pd.DataFrame()
clin = pd.read_csv('data/clindata.csv', sep = ';')
ZNF99mut = []
MAGEC1mut = []
patients = []
for ever in clin['PATIENT_ID']:
    patients.append(ever)
    if ever in ZNF99names:
        ZNF99mut.append('mut')
    else:
        ZNF99mut.append('norm')
    if ever in MAGEC1names:
        MAGEC1mut.append('mut')
    else:
        MAGEC1mut.append('norm')
meta['patients'] = patients
#meta['ZNF99mut'] = ZNF99mut
meta['MAGEC1mut'] = MAGEC1mut
meta.head()
del maf#   
```
